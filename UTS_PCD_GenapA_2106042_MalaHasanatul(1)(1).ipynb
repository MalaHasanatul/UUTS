```python
# Histogram Array

import numpy as np
a = np.array([[1,2,2,2,10],[10,20,30,40,50]])
print(a)
print(a.ravel())
print(a.flatten())
plt.hist(a.ravel(),bins = 50, range = [0,50])
plt.show()

    # Menciptakan array NumPy a.
    # a.ravel(): Mengubah array 2D menjadi array 1D.
    # a.flatten(): Sama seperti ravel(), menghasilkan array 1D.
    # plt.hist(a.ravel(), bins=50, range=[0,50]): Membuat histogram dari array a.
```

```python
# Histogram Warna untuk Setiap Saluran

color = ('b','g','r')
for i,col in enumerate(color):
  histogr = cv.calcHist([image], [i], None, [256], [0,256])
  plt.plot(histogr, color = col)
  plt.xlim([0,325])
plt.show()

# Membuat dan menampilkan histogram untuk setiap saluran warna (biru, hijau, merah) pada gambar.
```

```python
# Konversi Gambar ke Grayscale

gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)

    # cv.cvtColor(image, cv.COLOR_BGR2GRAY): Mengubah gambar warna menjadi gambar keabuan.
    # cv2_imshow(gray_image): Menampilkan gambar keabuan.
```

```python
# Histogram untuk Gambar Keabuan

plt.hist(gray_image.ravel(), bins = 256, range = [0,256])
plt.show()
# Membuat dan menampilkan histogram untuk gambar keabuan dengan 256 bins.

print(gray_image.size)
print(gray_image.shape)
print(gray_image.dtype)

    # gray_image.size: Menampilkan ukuran total gambar keabuan.
    # gray_image.shape: Menampilkan dimensi gambar keabuan.
    # gray_image.dtype: Menampilkan tipe data piksel dalam gambar keabuan.
```

```python
# Histogram untuk Gambar Keabuan dengan Jumlah Bins yang Berbeda

plt.hist(gray_image.ravel(), bins = 4, range = [0,256])
plt.show()
# Sama seperti sebelumnya, tetapi dengan jumlah bins yang berbeda (4 bins)

# . Informasi tentang Gambar Keabuan (Lagi)
print(gray_image.size)
print(gray_image.shape)
print(gray_image.dtype)

# Sama seperti sebelumnya, memberikan informasi tentang gambar keabuan setelah manipulasi histogram
```

```python
# Inversi Gambar Keabuan

invert_gray = 255 - gray_image
cv2_imshow(invert_gray)

    # invert_gray = 255 - gray_image: Menghasilkan gambar keabuan yang terinvert.
    # cv2_imshow(invert_gray): Menampilkan gambar keabuan yang terinvert.
```

```python
# Histogram untuk Gambar Keabuan Terinvert

plt.hist(invert_gray.ravel(), bins = 256, range = [0,256])
plt.show()

# Sama seperti sebelumnya, tetapi untuk gambar keabuan yang terinvert
```

```python
# Pencerahan Gambar Keabuan

bright_image = (100.0/255)*gray_image + 100
cv2_imshow(bright_image)

    # bright_image = (100.0/255)*gray_image + 100: Menghasilkan gambar keabuan yang dicerahkan.
    # cv2_imshow(bright_image): Menampilkan gambar keabuan yang dicerahkan.
```

```python
# Histogram untuk Gambar Keabuan yang Dicerahkan

plt.hist(bright_image.ravel(), bins = 256, range = [0,256])
plt.show()

# Sama seperti sebelumnya, tetapi untuk gambar keabuan yang dicerahkan.
```

```python
# Peningkatan Kontras pada Gambar Keabuan

darken_image = 255.0*(gray_image/255.0)**2
cv2_imshow(darken_image)

    # darken_image = 255.0*(gray_image/255.0)**2: Menghasilkan gambar keabuan dengan kontras yang ditingkatkan.
    # cv2_imshow(darken_image): Menampilkan gambar keabuan dengan kontras yang ditingkatkan.
```

```python
# Histogram untuk Gambar Keabuan dengan Kontras yang Ditingkatkan

plt.hist(darken_image.ravel(), bins = 256, range = [0,256])
plt.show()

# Sama seperti sebelumnya, tetapi untuk gambar keabuan dengan kontras yang ditingkatkan.
```
